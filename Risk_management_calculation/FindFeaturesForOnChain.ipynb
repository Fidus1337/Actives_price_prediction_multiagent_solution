```python
def core(prefix: str) -> pd.DataFrame:
    suffixes = ("__lag1", "__lag3", "__lag5", "__lag7", "__lag15", "__diff1", "__pct1")
    cols = [c for c in df.columns if c.startswith(prefix) and not c.endswith(suffixes)]
    return df[["date"] + cols]

reserve_risk = core("index_btc_reserve_risk__")
active_addr  = core("index_btc_active_addresses__")
lth_supply   = core("index_btc_lth_supply__")
sth_supply   = core("index_btc_sth_supply__")
```
